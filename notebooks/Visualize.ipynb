## Get data
Import `.npy` data files for input, ground truth and prediction

```python
# Parameters
width = 256
height = 256
slices = 20
channels = 4
train_path = '../npy_data/train/'
test_path = '../npy_data/test/'

params = {'IMG_WIDTH' : width,
          'IMG_HEIGHT' : height,
          'NUM_SLICES' : slices,
          'IMG_CHANNELS' : channels,
          'TRAIN_PATH' : train_path,
          'TEST_PATH' : test_path}

(X_train, X_test, y_train, y_test) = get_data(**params)


preds_train = np.load(train_path+'preds/preds_train_'+str(channels)+'ch.npy')
preds_test = np.load(test_path+'preds/preds_test_'+str(channels)+'ch.npy')
preds_train_t = np.load(train_path+'preds/preds_train_t_'+str(channels)+'ch.npy')
preds_test_t = np.load(test_path+'preds/preds_test_t_'+str(channels)+'ch.npy')
```

```python
fig, ax = plt.subplots(1, 4, figsize=(20, 10), dpi=300)
for ix in range(4):
    ax[ix].imshow(X_train[10, ..., ix], cmap='gray')
    ax[ix].set_title('Channel '+str(ix+1))
plt.savefig('channels.pdf', dpi=300)
```
